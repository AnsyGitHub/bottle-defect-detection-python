```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
def underfilled(img):
  #converting into grayscale
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

  cv2_imshow(img)
  #cropping a part above the label  
  cropped = gray[150:165, 115:225]
  cv2_imshow(cropped) 
  np_cropped = np.array(cropped)

  #finding total pixels
  pixels = np_cropped.shape
  size = np_cropped.shape

  #thresholding
  thresh = np_cropped < 100
  summation = thresh.sum()

  #finding percentage of black pixels (<100) in the cropped image
  check = (summation / (size[0] * size[1]))*100
  if check > 25:
    print("The Bottle is filled")
  else:
    print("Underfilled")

#calling the function
img = cv2.imread("im.jpg")
underfilled(img)

img2 = cv2.imread("image004.jpg")
underfilled(img2)
```

```python
def overfilled(img):
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  cropped = gray[65:165, 160:200]
  cv2_imshow(img)
  cv2_imshow(cropped)
  np_cropped = np.array(cropped)
  size = np_cropped.shape
  thresh = np_cropped < 50
  summation = thresh.sum()
  check = (summation / (size[0] * size[1]))*100
  if check > 25:
    print("Overfilled")
  else:
    print("Filled")

image = cv2.imread('image004.jpg')
overfilled(image)
```

```python
def check_cap(image):
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  cv2_imshow(image)
  cropped = gray[10:65,150:200]
  cv2_imshow(cropped)
  np_cropped = np.array(cropped)
  size = np_cropped.shape
  thresh = np_cropped > 150
  summation = thresh.sum()
  check = (summation / (size[0] * size[1]))*100
  if check > 80:
      print("No Cap")
  else:
      print("Cap")

img = cv2.imread("image004.jpg")
check_cap(img)
print("\n")
img2 = cv2.imread("image008.jpg")
check_cap(img2)
```
